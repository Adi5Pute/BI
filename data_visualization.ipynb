```python
import pandas as pd

df = pd.read_csv('/content/sample_data/sample_dataset(2).csv')

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn

plt.figure(figsize=(8,5))
sbn.barplot(x='Name', y='Marks', data=df)
plt.title('Bar Plot (Name vs Marks)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sbn.lineplot(x='Age', y='Marks', data=df, marker='o')
plt.title('Line Plot (Age vs Marks)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sbn.histplot(df['Marks'], bins=4, kde=False)
plt.title('Histogram (Marks)')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sbn.scatterplot(x='Age',y='Marks',hue='Gender',data=df)
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sbn.boxplot(x='Gender',y='Marks',data=df)
plt.title("Box plot")
plt.show()
```
